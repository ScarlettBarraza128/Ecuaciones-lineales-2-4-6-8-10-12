#### Problema 2
$y - 5 = 12$Planteamiento: Queremos aislar la $y$.Despeje: El 5 que está restando pasa al lado derecho sumando.Solución: $y = 12 + 5 \Rightarrow y = 17$

```python
# Problema: y - 5 = 12
# Despeje: y = 12 + 5
# valor_derecha = 12
# valor_restado = 5
valor_derecha=int(input("Ingrese el valor de la derecha: "))
valor_restado=int(input("Ingrese el valor a restar: "))

y = valor_derecha + valor_restado
print(f"Resultado Problema 2: y = {y}")
```

#### Problema 4
$12 = m + 3m$Planteamiento: Combinamos términos semejantes en la derecha ($1m + 3m = 4m$).Ecuación: $12 = 4m$Despeje: El 4 que multiplica a $m$ pasa dividiendo al 12.Solución: $m = \frac{12}{4} \Rightarrow m = 3$

```python
# Problema: 12 = m + 3m
# Simplificación: 12 = 4m -> m = 12 / 4
# total = 12
# coeficiente_m = 1 + 3
total = int(input("Ingrese el total: "))
coeficiente_m = int(input("Ingrese el coeficiente de m: "))

m = total / coeficiente_m
print(f"Resultado Problema 4: m = {int(m)}")
```

#### Problema 6
 $5b + 2b - 3b = 16$Planteamiento: Operamos los coeficientes de $b$: $(5 + 2 - 3) = 4$.Ecuación: $4b = 16$Despeje: El 4 pasa dividiendo al 16.Solución: $b = \frac{16}{4} \Rightarrow b = 4$

```python
# Problema: 5b + 2b - 3b = 16
# Simplificación: 4b = 16 -> b = 16 / 4
# resultado_ecuacion = 16
# suma_coeficientes = 5 + 2 - 3
resultado_ecuacion = int(input("Ingrese el resultado de la ecuación: "))
suma_coeficientes = int(input("Ingrese la suma de los coeficientes: "))

b = resultado_ecuacion / suma_coeficientes
print(f"Resultado Problema 6: b = {int(b)}")
```

#### Problema 8
$7y - 5 = 3y + 7$Planteamiento: Movemos las $y$ a la izquierda y los números a la derecha: $7y - 3y = 7 + 5$.Ecuación: $4y = 12$Despeje: $y = \frac{12}{4}$Solución: $y = 3$

```python
# Problema: 7y - 5 = 3y + 7
# Agrupamos: (7 - 3)y = 7 + 5 -> 4y = 12
# terminos_independientes = 7 + 5
# coeficientes_y = 7 - 3
terminos_independientes = int(input("Ingrese los terminos independientes: "))
coeficientes_y = int(input("Ingrese los coeficientes de y: "))

y = terminos_independientes / coeficientes_y
print(f"Resultado Problema 8: y = {int(y)}")
```

#### Problema 10
$8n - 6 = 4n + 10$Planteamiento: Agrupamos: $8n - 4n = 10 + 6$.Ecuación: $4n = 16$Despeje: $n = \frac{16}{4}$Solución: $n = 4$

```python
# Problema: 8n - 6 = 4n + 10
# Agrupamos: (8 - 4)n = 10 + 6 -> 4n = 16
# lado_derecho = 10 + 6
# lado_izquierdo_n = 8 - 4
lado_derecho = int(input("Ingrese el lado derecho: "))
lado_izquierdo_n = int(input("Ingrese el lado izquierdo de n: "))
n = lado_derecho / lado_izquierdo_n
print(f"Resultado Problema 10: n = {int(n)}")
```

#### Problema 12
$10b - 8 = 4b + 10$Planteamiento: Agrupamos: $10b - 4b = 10 + 8$.Ecuación: $6b = 18$Despeje: $b = \frac{18}{6}$Solución: $b = 3$

```python
# Problema: 10b - 8 = 4b + 10
# Agrupamos: (10 - 4)b = 10 + 8 -> 6b = 18
# suma_numeros = 10 + 8
# resta_coeficientes = 10 - 4
suma_numeros = int(input("Ingrese la suma de los numeros: "))
resta_coeficientes = int(input("Ingrese la resta de los coeficientes: "))
b = suma_numeros / resta_coeficientes
print(f"Resultado Problema 12: b = {int(b)}")
```
